```python
# Librerias para tokenizar un texto y para tener una lista de stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopEnglish=set(stopwords.words('english'))
stopSpanish=set(stopwords.words('spanish'))
```

```python
stopEnglish
```

```python
stopSpanish
```

```python
all_stop = stopEnglish.union(stopSpanish)
```

```python
all_stop
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
q = questions[0]
```

```python
q
```

```python
data[q]
```

```python
", ".join(['1','2','3'])
```

```python
", ".join(['1',2,'3'])
```

```python
" ".join(data[q].astype(str))
```

```python
text = " ".join(data[q].astype(str))
text = word_tokenize(text)
```

```python
text = [word for word in text if word not in all_stop]
text
```

```python
text = " ".join(text)
text
```

```python
wordcloud = WordCloud(width=800, height=400).generate(text)
# Display the generated image:
# the matplotlib way: 
plt.figure( figsize=(15,8) )
plt.imshow(wordcloud) 
plt.axis("off")
plt.show()
```

```python
for q in questions:
  print("--------------------------------")
  print(q)
  print("///////////////////////////Columna de dataframe///////////////////////////")
  print(data[q])
  text = " ".join(data[q].astype(str))
  print("///////////////////////////Texto concatenado///////////////////////////")
  print(text)
  print("///////////////////////////Texto tokenizado///////////////////////////")
  text = word_tokenize(text)
  print(text)
  text = [word for word in text if word not in all_stop]
  print("///////////////////////////Texto tokenizado filtrado///////////////////////////")
  print(text)
  print("///////////////////////////Texto filtrado concatenado///////////////////////////")
  text = " ".join(text)
  print(text)
  # Generate a word cloud image
  wordcloud = WordCloud(width=800, height=400).generate(text)
  # Display the generated image:
  # the matplotlib way: 
  plt.figure( figsize=(15,8) )
  plt.imshow(wordcloud) 
  plt.axis("off")
  plt.show()
```
