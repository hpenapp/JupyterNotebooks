## K - means

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11): 
     kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300,n_init=10,random_state=0 )
     kmeans.fit(df)
     wcss.append(kmeans.inertia_)
```

Se hace el metodo grafica, para determinar la mejor cantidad de clusters 

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Gracias a la grafica, podemos determinar que hay 6 grupos de clusters

```python
kmeans = KMeans(n_clusters=6, init ='k-means++',max_iter=300,n_init=10,random_state=0 )
y_kmeans = kmeans.fit_predict(df)
```

```python
plt.scatter(df[y_kmeans==0, 0], df[y_kmeans==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(df[y_kmeans==1, 0], df[y_kmeans==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(df[y_kmeans==2, 0], df[y_kmeans==2, 1], s=100, c='green', label ='Cluster 3')
plt.scatter(df[y_kmeans==3, 0], df[y_kmeans==3, 1], s=100, c='cyan', label ='Cluster 4')
plt.scatter(df[y_kmeans==4, 0], df[y_kmeans==4, 1], s=100, c='magenta', label ='Cluster 5')
plt.scatter(df[y_kmeans==5, 0], df[y_kmeans==5, 1], s=100, c='orange', label ='Cluster 6')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label ='Centroids')
plt.title('Clusters de frutas')
plt.show()
```

Como podemos ver, se observa que hay 6 tipos diferentes de frutas segun este algoritmo, y como podemos ver, el grupo que esta en verde es el tipo de fruta que predomina, y le sigue en naranja y asi
