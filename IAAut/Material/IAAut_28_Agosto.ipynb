```python
def g_m(x,y,m,b):
  # Calcular gradiente de J(m,b) de m evaluado en m,b 
  # usando los datos x,y
  return np.mean(-2*(y-m*x-b)*x)


def g_b(x,y,m,b):
  # Calcular gradiente de J(m,b) de b evaluado en m,b 
  # usando los datos x,y
  return np.mean(-2*(y-m*x-b))
```

```python
# Inicializar los parametros aleatoriamente
m = np.random.randint(0,10)
b = np.random.randint(0,10)
n_iter = 1000
k = 0 
lr = 0.1
batch = 20
tol = 0.01
while k<=n_iter:
  Jant = J(m,b,x,y)
  # Calcular gradientes (sobre un subconjunto de datos aleatorio) 
  sub = np.random.randint(low=0,high=N,size=batch)
  gr_b = g_b(x[sub,:],y[sub,:],m,b)
  gr_m = g_m(x[sub,:],y[sub,:],m,b)
  # Mover los parametros en la direccion opuesta del gradiente
  m = m- lr*gr_m
  b = b- lr*gr_b
  Jnew = J(m,b,x,y)
  if np.abs(Jnew-Jant) <= tol:
    k = n_iter
  k = k + 1
m,b, J(m,b,x,y)
```
